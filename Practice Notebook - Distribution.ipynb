```python
import nsfg
import matplotlib.pyplot as plt
```

```python
preg = nsfg.ReadFemPreg()
preg.columns
```

```python
live = preg[preg.outcome ==1]
#preg['outcome'].value_counts().sort_index().plot(kind = 'bar')
x = live['birthwgt_lb'].value_counts().index
y = live['birthwgt_lb'].value_counts()
plt.xlabel('pounds')
plt.ylabel('frequency')
plt.bar(x,y,label = 'birthwgt_lb')
plt.legend()
```

```python
import thinkplot
import thinkstats2

hist = thinkstats2.Hist(live.birthwgt_lb, label='birthwgt_lb')
thinkplot.Hist(hist)
thinkplot.Show(xlabel='pounds', ylabel='frequency')
```

```python
fig,ax = plt.subplots(2,2,figsize=(15,9))
x1 = live['birthwgt_lb'].value_counts().index
y1 = live['birthwgt_lb'].value_counts()
x2 = live['birthwgt_oz'].value_counts().index
y2 = live['birthwgt_oz'].value_counts()
x3 = live['agepreg'].value_counts().index
y3 = live['agepreg'].value_counts()
x4 = live['prglngth'].value_counts().index
y4 = live['prglngth'].value_counts()

ax[0,0].bar(x1,fig,ax = plt.subplots(2,2,figsize=(15,9))
ax[0,0].legend()
ax[0,0].set_xlabel('pounds')
ax[0,0].set_ylabel('frequency')


ax[0,1].bar(x2,y2,width=0.5,color='g',label='birthwgt_oz')
ax[0,1].legend()
ax[0,1].set_xlabel('ounces')
ax[0,1].set_ylabel('frequency')

ax[1,0].bar(x3,y3,width=0.5,color="g",label='agepreg')
ax[1,0].legend()
ax[1,0].set_xlabel('years')
ax[1,0].set_ylabel('frequency')

ax[1,1].bar(x4,y4,width=0.5,color="g",label='prglngth')
ax[1,1].legend()
ax[1,1].set_xlabel('weeks')
ax[1,1].set_ylabel('frequency')
```

```python
print(y4.sort_index().nlargest(10))
print(y4.sort_index().nsmallest(10))
```

```python
# first baby, only focus on preg lengh > 27 weeks, and live birth
firsts = live[live.birthord == 1]
others = live[live.birthord != 1]
x1 = firsts['prglngth'].value_counts().index
y1 = firsts['prglngth'].value_counts()
x2 = others['prglngth'].value_counts().index
y2 = others['prglngth'].value_counts()
w=0.45

plt.bar(x1,y1,width = w,color ='g',label='firsts',align='center')
plt.bar(x2+w,y2,width = w,color ='r',label='others',align='center')
plt.xlim(27,46)
plt.legend()
plt.xlabel('weeks')
plt.ylabel('frequency')
```
